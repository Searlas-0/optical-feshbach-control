```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt


def scat_len(intensity, detuning, loss=1):
    # l* = a_bg 
    loss = float(loss)
    I = jnp.asarray(intensity, dtype=jnp.float32) 
    v = jnp.asarray(detuning, dtype=jnp.float32) 

    a_s = 1.0 + (I / (-v - I + loss * 0.5j))

    return a_s 




```
